```python
raw_2019 = pd.read_csv('./rawfiles/SUBWAY_TIME_2019.csv')
raw_2020 = pd.read_csv('./rawfiles/SUBWAY_TIME_2020.csv')
raw_2021 = pd.read_csv('./rawfiles/SUBWAY_TIME_2021.csv')
raw_2022= pd.read_csv('./rawfiles/SUBWAY_TIME_2022.csv')
raw_list=[raw_2019,raw_2020,raw_2021,raw_2022]
raw_2019
```

```python
for raw in raw_list:
    raw['사용월']= pd.to_datetime(raw['사용월'],format='%Y%m')
    raw['연월'] = raw['사용월'].dt.strftime('%Y-%m')

```

```python
raw_2019.head()
```

```python
for a in raw_list:
    a["출근시간"]=a['07시-08시 승차인원']+a['08시-09시 승차인원']+a['09시-10시 승차인원']
    a["퇴근시간"]=a['17시-18시 승차인원']+a['18시-19시 승차인원']+a['19시-20시 승차인원']
```

```python
raw_2019
```

```python

night2019 = raw_2019.groupby('연월')[['호선명','21시-22시 승차인원','22시-23시 승차인원','23시-24시 승차인원','00시-01시 승차인원','01시-02시 승차인원']].sum()
night2019=night2019.reset_index()

night2020 = raw_2020.groupby('연월')[['호선명','21시-22시 승차인원','22시-23시 승차인원','23시-24시 승차인원','00시-01시 승차인원','01시-02시 승차인원']].sum()
night2020=night2020.reset_index()

night2021 = raw_2021.groupby('연월')[['호선명','21시-22시 승차인원','22시-23시 승차인원','23시-24시 승차인원','00시-01시 승차인원','01시-02시 승차인원']].sum()
night2021=night2021.reset_index()

night2022 = raw_2022.groupby('연월')[['호선명','21시-22시 승차인원','22시-23시 승차인원','23시-24시 승차인원','00시-01시 승차인원','01시-02시 승차인원']].sum()
night2022=night2022.reset_index()
```

```python
for i in [night2019,night2020,night2021,night2022]:
    i["총승차승객"]=i.sum(axis= 1,numeric_only= True)
    i=i[['연월','총승차승객']]
```
